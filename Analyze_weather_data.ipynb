```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_data = pd.read_csv("/content/Temp_and_rain.csv")
```

```python
print(weather_data.describe())
```

```python
plt.hist(weather_data["Temperature"])
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Number of days")
plt.title("Distribution of temperature in weather data")
plt.show()
```

```python
correlation = weather_data["Temperature"].corr(weather_data["rain"])
print("The correlation between temperature and rain is:", correlation)
```

```python
plt.scatter(weather_data["Temperature"], weather_data["rain"])
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Rain (mm)")
plt.title("Relationship between temperature and rain in weather data")
plt.show()
```

```python
monthly_average_temperature = weather_data.groupby("Month")["Temperature"].mean()
print(monthly_average_temperature)
```

```python
annual_average_rain = weather_data.groupby("Year")["rain"].mean()
print(annual_average_rain)
```
